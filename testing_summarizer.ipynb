```python
import os
from dotenv import load_dotenv
from langchain.chains import load_summarize_chain
from langchain.document_loaders import YoutubeLoader, UnstructuredURLLoader
from langchain_groq.chat_models import ChatGroq
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema import Document

load_dotenv()
```

```python
#https://www.youtube.com/watch?v=4qcgqOCDv2c
```

```python
from youtube_transcript_api import YouTubeTranscriptApi

video_id = "4qcgqOCDv2c"#"g-_lw944edU&t=1018s"#"F8NKVhkZZWI&t=1s"#"qYNweeDHiyU"
transcript = YouTubeTranscriptApi.get_transcript(video_id)

# Join transcript chunks into text
text = "\n".join([entry['text'] for entry in transcript])
print(text)
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap=100)
splitted_docs = text_splitter.split_text(text)
doc = [Document(page_content=chunk) for chunk in splitted_docs]
```

```python
doc
```

```python
from langchain.prompts import PromptTemplate
groq_api_key = os.getenv("GROQ_API_KEY")
llm = ChatGroq(model="llama-3.2-3b-preview", groq_api_key=groq_api_key)
prompt_template = """
Provide a detailed summary of the following content in 500 words:
Content: {text}
"""
prompt = PromptTemplate(template=prompt_template, input_variables=["text"])
```

```python
chain = load_summarize_chain(llm, chain_type="stuff", prompt=prompt)
output_summary = chain.run(doc)
```

```python
print(output_summary)
```
